## 1. Physics

```python
# IQR = Q3 - Q1
Q3 = df['Phy_marks'].quantile(0.75)
Q1 = df['Phy_marks'].quantile(0.25)

IQR = Q3 - Q1

upper_limit = Q3 + (IQR * 1.5)

lower_limit = Q1 - (IQR * 1.5)
```

```python
Q3
```

```python
Q1
```

```python
IQR
```

```python
upper_limit
```

```python
lower_limit
```

## Removing the Outliers

```python
phy_outliers = np.where(df['Phy_marks'] > upper_limit, True, np.where(df['Phy_marks'] < lower_limit, True, False))
phy_outliers
```

```python
phy_having_outliers = df.loc[(phy_outliers), ]
phy_having_outliers
```

```python
df = df.drop(phy_having_outliers.index)
```

# Normalize

```python
# Range converts to: 0 to 1
def norm_func(i): 
    x = (i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm = norm_func(df['Percentage'])
df_norm
```
